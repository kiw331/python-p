```python
#[완료6장_함수_책코드] 텍스트프로그램 함수로 완성
def select_menu(menu):
    order = {}
  
    while True:
        item = input("메뉴 이름을 입력하세요 (종료는 exit): ")
        if item == "exit":
            break
        elif item not in menu:
            print("잘못된 입력입니다.")
            continue

        count = input("수량을 입력하세요: ")
        count = int(count)
        order[item] = count    
    return order


def print_order(menu, order):
    total_price = 0
    print("주문 내역:")
    for item, count in order.items():
        price = menu[item]
        order_price = price * count
        total_price += order_price
        print(f" - {item} ({price}원) x {count} = {order_price}원")
    print(f"\n총 가격: {total_price}원")
    return total_price


pizza_menu = {
    "페퍼로니 피자": 3000,
    "치즈 피자": 3200,
    "콤비네이션 피자": 3500,
    "불고기 피자": 3600,
    "해산물 피자": 3800
}
drink_menu = {
    "콜라": 1500,
    "사이다": 1500,
    "생수": 1000
}

print("피자를 선택해주세요.")
for pizza, price in pizza_menu.items():
        print(f"{pizza} ({price}원)")

pizza_order = select_menu(pizza_menu)

print("음료수를 선택해주세요.")
for drink, price in drink_menu.items():
    print(f"{drink} ({price}원)")
drink_order = select_menu(drink_menu)

pizza_total_price=print_order(pizza_menu, pizza_order)
drink_total_price=print_order(drink_menu, drink_order)
print("************************************************************")
print("피자와 음료 총 가격:", pizza_total_price + drink_total_price)
```

```python
#[완료9장_클래스_1_책코드)

class Menu:
    def __init__(self, menu_dict):
        self.menu = menu_dict

    def select(self):
        order = {}
        while True:
            item = input("메뉴 이름을 입력하세요 (종료는 exit): ")
            if item == "exit":
                break
            elif item not in self.menu:
                print("잘못된 입력입니다.")
                continue

            count = input("수량을 입력하세요: ")
            count = int(count)            
            order[item] = count

        return order

    def print_order(self, order):
        total_price = 0
        print("주문 내역:")
        for item, count in order.items():
            price = self.menu[item]
            order_price = price * count
            total_price += order_price
            print(f" - {item} ({price}원) x {count} = {order_price}원")
        print(f"\n총 가격: {total_price}원")
        return total_price


pizza_menu = Menu({
    "페퍼로니 피자": 3000,
    "치즈 피자": 3200,
    "콤비네이션 피자": 3500,
    "불고기 피자": 3600,
    "해산물 피자": 3800
})
drink_menu = Menu({
    "콜라": 1500,
    "사이다": 1500,
    "생수": 1000
})

print("피자를 선택해주세요.")
for pizza, price in pizza_menu.menu.items():
    print(f"{pizza} ({price}원)")

pizza_order = pizza_menu.select()

print("음료수를 선택해주세요.")
for drink, price in drink_menu.menu.items():
    print(f"{drink} ({price}원)")
drink_order = drink_menu.select()

pizza_total_price = pizza_menu.print_order(pizza_order)
drink_total_price = drink_menu.print_order(drink_order)

print("************************************************************")
print("피자와 음료 총 가격:", pizza_total_price + drink_total_price)
```

```python
#[완료12장tkinter_3]tkinter 클래스로 완성_주문메뉴만출력
#tkinter클래스로 피자주문 완성 코드

import tkinter as tk

class PizzaOrderApp:
    def __init__(self, root):
        self.root = root
        root.title("조각 피자 주문 프로그램")

        # 피자 메뉴와 가격
        self.pizza_menu = {
            "페퍼로니 피자": 3000,
            "치즈 피자": 3200,
            "콤비네이션 피자": 3500,
            "불고기 피자": 3600,
            "해산물 피자": 3800
        }

        # 음료수 메뉴와 가격
        self.drink_menu = {
            "콜라": 1500,
            "사이다": 1500,
            "생수": 1000
        }

        # 피자 선택 엔트리
        self.pizza_vars = {}
        self.pizza_frame = tk.LabelFrame(root, text="피자")
        self.pizza_frame.pack(side='left', padx=10, pady=10, fill="both", \
                              expand="yes")
        for pizza, price in self.pizza_menu.items():
            self.pizza_vars[pizza] = tk.StringVar(value="0")
            lbl = tk.Label(self.pizza_frame, text=f"{pizza} ({price}원)")
            lbl.pack(anchor="w")
            ent = tk.Entry(self.pizza_frame, textvariable=self.pizza_vars\
                           [pizza])
            ent.pack(anchor="w")

        # 음료수 선택 엔트리
        self.drink_vars = {}
        self.drink_frame = tk.LabelFrame(root, text="음료수")
        self.drink_frame.pack(side='right', padx=10, pady=10, fill="both",\
                              expand="yes")
        for drink, price in self.drink_menu.items():
            self.drink_vars[drink] = tk.StringVar(value="0")
            lbl = tk.Label(self.drink_frame, text=f"{drink} ({price}원)")
            lbl.pack(anchor="e")
            ent = tk.Entry(self.drink_frame, textvariable=self.drink_vars[drink])
            ent.pack(anchor="e")

        # 주문 버튼
        btn = tk.Button(root, text="주문", command=self.order)
        btn.pack(pady=10)

        # 주문내역 출력 레이블
        self.order_text = tk.StringVar()
        self.order_text.set("주문내역:\n")
        order_label = tk.Label(root, textvariable=self.order_text, justify=\
                               "left")
        order_label.pack(padx=10, pady=10, fill="both", expand="yes")
    def order(self):
        total_price = 0
        order_list = "주문내역:\n"

        # 선택한 피자 메뉴와 가격 출력
        order_list += "피자:\n"
        for pizza, price in self.pizza_menu.items():
            if self.pizza_vars[pizza].get():
                pizza_count = int(self.pizza_vars[pizza].get())
                if pizza_count > 0:
                    order_list += f" - {pizza} ({price}원) x {pizza_count}\n"
                    total_price += price * pizza_count
                    
                    
                    
        order_list += "음료수:\n"
        for drink, price in self.drink_menu.items():
            if int(self.drink_vars[drink].get()):
                drink_count = int(self.drink_vars[drink].get())
                if drink_count > 0:
                    order_list += f" - {drink} ({price}원) x {drink_count}\n"
                    total_price += price * drink_count


        # 주문내역 출력
        order_list += f"\n총 가격: {total_price}원"
        self.order_text.set(order_list)
        
if __name__ == '__main__':
    root = tk.Tk()
    app = PizzaOrderApp(root)
    root.mainloop()
    
```
