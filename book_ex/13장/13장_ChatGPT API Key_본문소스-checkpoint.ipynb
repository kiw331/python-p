```python
#ㅔ360 소스
import openai
import time

# OpenAI API key
openai.api_key = "YOUR_API_KEY"

# GPT-3 생성 모델 ID
model_engine = "text-davinci-002"

# ChatGPT 상호 작용 함수
def chat_with_gpt(prompt):
    response = openai.Completion.create(
        engine=model_engine,
        prompt=prompt,
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.5,
    )

    message = response.choices[0].text.strip()
    return message

# ChatGPT에 인사하기
print(chat_with_gpt("안녕하세요!"))

# ChatGPT와 대화하기
while True:
    user_input = input("당신: ")
    response = chat_with_gpt("나: " + user_input)
    print("ChatGPT: " + response)
    time.sleep(1)
```

```python
#p366 소스
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Self-esteem"
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')
content = soup.find(id='mw-content-text').find(class_='mw-parser-output')

# 제목 가져오기
title = soup.find(id="firstHeading").text

# 본문 가져오기
paragraphs = content.find_all('p')
text = ''
for p in paragraphs:
    text += p.text

print(title)
print(text)
```

```python
#p369 소스
from pptx import Presentation
from pptx.util import Inches

# 새로운 프레젠테이션 생성
prs = Presentation()

# 제목 슬라이드 생성
slide = prs.slides.add_slide(prs.slide_layouts[0])
title = slide.shapes.title
title.text = "파워포인트 제목"

# 내용 슬라이드 생성
slide = prs.slides.add_slide(prs.slide_layouts[1])
title = slide.shapes.title
title.text = "내용 제목"
body = slide.shapes.placeholders[1]
tf = body.text_frame
tf.text = "내용 텍스트 추가"
tf.add_paragraph().text = "새로운 단락 추가"

# 이미지 슬라이드 생성
slide = prs.slides.add_slide(prs.slide_layouts[1])
title = slide.shapes.title
title.text = "이미지 제목"
left = Inches(1)
top = Inches(2)
height = Inches(5.5)
pic = slide.shapes.add_picture("image.jpg", left, top, height=height)

# 파워포인트 파일 저장
prs.save("test.pptx")
```
